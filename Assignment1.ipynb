```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(-15,16,1)
y=np.arange(-15,16,1)
print(x,y)
```

```python
plt.style.use('seaborn-whitegrid')

plt.axes()
plt.plot(x,y)
plt.show()
```

```python
np.meshgrid(x,y)
```

```python
# Create a grid of points in x-y space 
xvals = np.arange(-15, 16, 1)
yvals = np.arange(-15, 16, 1)
xygridPoints = np.column_stack([[x, y] for x in xvals for y in yvals])
print(xygridPoints)
```

```python
# Apply linear transform
a = np.column_stack([[1, 0], [1, 1]])
print(a)
uvgrid = np.dot(a, xygridPoints)
print(uvgrid)
```

```python
# Plot grid points 
plt.figure(figsize=(10, 10), facecolor="w")
plt.scatter(xygridPoints[0], xygridPoints[1], s=36, edgecolor="none")
# Set axis limits
plt.grid(True)
plt.axis("equal")
plt.title("Original grid in x-y space")
```

```python
# Plot transformed grid points
plt.figure(figsize=(10, 10), facecolor="w")
plt.scatter(uvgrid[0], uvgrid[1], s=36, edgecolor="none")
plt.grid(True)
plt.axis("equal")
plt.title("Transformed grid ")
```
